```python
import re #модуль re
filename = "folder1/folder2/file.ext"
extension = re.search(r'\.(\w+)$', filename).group(1) #функция search() в модуле re находит подстроку, которая начинается с точки и заканчивается концом строки + содержит один или более символов (`\w+`)
print(extension)
```

Задание 2

```python
import http #пакет http
help(http)
from http import client
conn = client.HTTPSConnection("www.google.com") # создаем соединение с google
conn.request("GET", "/") #GET-запрос на сервер
response = conn.getresponse() #ответ на запрос
print(response.status, response.reason)
data = response.read()
print(data)
conn.close()
```
